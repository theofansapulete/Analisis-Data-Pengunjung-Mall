```python
#memasukan library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

%matplotlib inline
```

```python
data = pd.read_csv('Pengunjung_Mall.csv')
```

```python
print(data.head())  # Tampilkan beberapa baris pertama dataset
print(data.info())  # Tampilkan informasi dataset
```

```python
#data
data = pd.DataFrame({
    'Umur': [19, 21, 20, 23, 31],
    'Pendapatan (ribuan $)': [15, 15, 16, 16, 17]
})
```

```python
#umur
plt.figure(figsize=(10, 6))
sns.histplot(data['Umur'], bins=30, kde=True)
plt.title('Distribusi Umur Pengunjung Mall')
plt.xlabel('Umur')
plt.ylabel('Jumlah Pengunjung')
plt.grid(True)
plt.show()
```

```python
#jenis kelamin
plt.figure(figsize=(6, 6))
sns.countplot(data['Jenis Kelamin'])
plt.title('Jumlah Pengunjung Mall berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah Pengunjung')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1, 3, 3)
sns.scatterplot(x='Umur', y='Kemampuan Membeli (1-100)', data=data, hue='Jenis Kelamin', palette='Set2')
plt.title('Umur vs Kemampuan Membeli')
plt.tight_layout()
plt.show()
```
